# Telco Customer Churn

### 1. Setup

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Data Preparation

```python
# Load the dataset
data_utl = 'https://raw.githubusercontent.com/RajithaSampathV/Customer-Churn-Analysis/refs/heads/main/Telco_Customer_Churn_Dataset.csv'
df = pd.read_csv(data_utl)
```

```python
print("Shape:", df.shape)
print(df.isnull().sum())
df.head()
```

```python
# Quick schema overview
df.info()
```

```python
#convert to numeric
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
# Remove obvious identifier
df = df.drop(columns=['customerID'])
```

```python
for col in df.columns:
    print(f"{col}: {df[col].unique()}")
```

```python
# # Binary Encording
bin_enc = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']
for col in bin_enc:
    df[col] = df[col].map({'Yes': 1, 'No': 0})
```

```python
# Label Encoding
label  = LabelEncoder()
cat_cols = ['gender', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaymentMethod']
for col in cat_cols:
    df[col] = label.fit_transform(df[col])
```

```python
df.head()
```

```python
print(df.isnull().sum())
```

```python
# Check the predictive value count
df['Churn'].value_counts()
```

```python

#Drop the rows if churn value is 0 and TotalChurges is null
null_TotalCharges = df[df['TotalCharges'].isnull()]
null_TotalCharges['Churn'].value_counts()

df = df.drop(df[(df['TotalCharges'].isna()) & (df['Churn'] == 0)].index)
```
